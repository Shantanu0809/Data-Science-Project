```python
df_survey=pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
df_survey
```

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') 


print ('Matplotlib version: ', mpl.__version__)
```

```python
df_survey.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Votes') # add to x-label to the plot
plt.ylabel('Courses') # add y-label to the plot
plt.title("Percentage of Respondent's Interest in Data Science Areas") # add title to the plot

plt.show()
```

```python
df_survey = df_survey.sort_values('Very interested', ascending=False)
respondents = 2233
df_survey = df_survey / respondents * 100
df_survey.round(2)
```

```python
per = df_survey.plot(kind='bar', 
            figsize=(20, 8), 
            width=0.8, 
            color=['#5cb85c', '#5bc0de', '#d9534f'],
           )

per.set_title("Percentage of Respondents' Interest in Data Science Area", fontsize=16)
per.spines['left'].set_visible(False)
per.spines['top'].set_visible(False)
per.spines['right'].set_visible(False)
per.axes.get_yaxis().set_visible(False)
per.tick_params(labelsize=14)
per.legend(fontsize=14)
for p in per.patches:
    per.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )
plt.show()
```
